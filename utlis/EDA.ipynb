# Exploratory Data Analysis

## Total Data per Class

```python
import os
from collections import Counter

train_dir = "../data/WLBisindo/split/train"

counts = {}

for label in os.listdir(train_dir):
    label_path = os.path.join(train_dir, label)
    if os.path.isdir(label_path):
        counts[label] = len(os.listdir(label_path))

print(counts)
```

## Sample Frames

## Frames per Videos

```python
frame_counts = []

for label in os.listdir(train_dir):
    label_path = os.path.join(train_dir, label)

    for vid in os.listdir(label_path):
        vid_path = os.path.join(label_path, vid)
        frame_counts.append(len(os.listdir(vid_path)))

print("Min:", min(frame_counts))
print("Max:", max(frame_counts))
```

## Video Lengths Distribution

```python
import cv2
import os

raw_dir = "../data/WLBisindo/raw"

lengths = []

for file in os.listdir(raw_dir):
    path = os.path.join(raw_dir, file)
    cap = cv2.VideoCapture(path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    lengths.append(total)
    cap.release()

print("Min:", min(lengths))
print("Max:", max(lengths))
```

## Motion Analysis

```python
def motion_score(frame1, frame2):
    diff = cv2.absdiff(frame1, frame2)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    return gray.sum()
```

```python
import matplotlib.pyplot as plt

scores = []

# ambil 1 video sample
video_path = ...

cap = cv2.VideoCapture(video_path)
ret, prev = cap.read()

while True:
    ret, curr = cap.read()
    if not ret:
        break

    score = motion_score(prev, curr)
    scores.append(score)
    prev = curr

cap.release()

plt.plot(scores)
plt.title("Motion per Frame")
plt.show()
```

## Video Blur Analysis

```python
def blur_score(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    return cv2.Laplacian(gray, cv2.CV_64F).var()
```

```python
blur_scores = []

# loop beberapa frame dataset
for ...:
    blur_scores.append(blur_score(frame))

plt.hist(blur_scores, bins=50)
plt.title("Blur Distribution")
plt.show()
```
