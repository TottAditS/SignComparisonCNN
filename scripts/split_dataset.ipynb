## Frames Dataset Splitting

```python
import os
import random
import shutil
from tqdm import tqdm

BASE_DIR = os.path.abspath("..")
FRAMES_DIR = os.path.join(BASE_DIR, "data/WLBisindo/frames")
OUTPUT_DIR = os.path.join(BASE_DIR, "data/WLBisindo/split")

train_ratio = 0.7
val_ratio = 0.15

# hitung total video (biar progress bar akurat)
total_videos = 0
for label in os.listdir(FRAMES_DIR):
    label_path = os.path.join(FRAMES_DIR, label)
    if os.path.isdir(label_path):
        total_videos += len(os.listdir(label_path))

# progress bar utama
with tqdm(total=total_videos, desc="Splitting Dataset") as pbar:

    for label in os.listdir(FRAMES_DIR):
        label_path = os.path.join(FRAMES_DIR, label)

        if not os.path.isdir(label_path):
            continue

        videos = os.listdir(label_path)
        random.shuffle(videos)

        n = len(videos)
        n_train = int(n * train_ratio)
        n_val = int(n * val_ratio)

        train_videos = videos[:n_train]
        val_videos = videos[n_train:n_train+n_val]
        test_videos = videos[n_train+n_val:]

        for split_name, split_videos in zip(
            ["train", "val", "test"],
            [train_videos, val_videos, test_videos]
        ):
            for vid in split_videos:
                src = os.path.join(label_path, vid)
                dst = os.path.join(OUTPUT_DIR, split_name, label, vid)

                os.makedirs(os.path.dirname(dst), exist_ok=True)

                # copy folder video (isi frame)
                if not os.path.exists(dst):
                    shutil.copytree(src, dst)

                pbar.update(1)
```
